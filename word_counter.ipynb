```python
from bs4 import BeautifulSoup
import nltk
from nltk.corpus import stopwords
from collections import Counter
```

```python
r = BeautifulSoup(open('/kaggle/input/the-adventures-of-sherlock-holmes/pg1661-images.html', encoding='utf8'), 'html.parser')
```

```python
html = r.text
#print(html[0:3000])
```

```python
book_text = r.get_text()
```

```python
tokenizer = nltk.tokenize.RegexpTokenizer('\w+')
tokens = tokenizer.tokenize(book_text)
```

```python
words = [token.lower() for token in tokens]
words[:10]
```

```python
stop_words = nltk.corpus.stopwords.words('english')
stop_words[:10]
```

```python
words_no_stop = [word for word in words if word not in stop_words]
words_no_stop[:10]
```

```python
count = Counter(words_no_stop)
count_all = Counter(words)
```

```python
top = count.most_common(100)
print(top)
```

```python
top_all = count_all.most_common(100)
print(top_all)
```
